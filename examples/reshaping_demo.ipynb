# Reshaping and summarising

Visualise the operations that rearrange or summarise a tensor. Each view places the source tensor and the result side by side so the mapping is easy to read.

```python
from pprint import pprint

import numpy as np

import rainbow_tensor as rt
```

## Reshape

`reshape` keeps the row major order, so element k stays element k while the layout changes. The same values flow from the old shape into the new one.

```python
x = np.arange(6).reshape(2, 3)
pprint(x)
pprint(x.reshape(3, 2))
rt.reshape(x, (3, 2))
```

A single `-1` lets one axis be inferred from the rest, just like NumPy.

```python
x = np.arange(24).reshape(2, 3, 4)
pprint(x)
pprint(x.reshape(-1, 4))
rt.reshape(x, (-1, 4))
```

## Transpose and permute

`transpose` reorders the axes. Each axis keeps its colour as it moves, so you can trace where it lands. With no argument the axes are reversed.

```python
x = np.arange(6).reshape(2, 3)
pprint(x)
pprint(x.T)
rt.transpose(x)
```

Pass an explicit permutation to move the axes where you want them.

```python
x = np.arange(24).reshape(2, 3, 4)
pprint(x)
pprint(x.transpose(2, 0, 1))
rt.transpose(x, (2, 0, 1))
```

## Axis reductions

`sum` and `mean` collapse one axis. Source values that fold into the same result share one background, the first group is highlighted, and the result panel shows the surviving shape.

```python
x = np.arange(12).reshape(3, 4)
pprint(x)
pprint(x.sum(axis=0))
rt.sum(x, 0)
```

`mean` divides each group by its count, so the result holds floats.

```python
x = np.arange(12).reshape(3, 4)
pprint(x)
pprint(x.mean(axis=1))
rt.mean(x, 1)
```

## Dark theme

Every view also renders in the dark theme through `theme="dark"`.

```python
x = np.arange(12).reshape(3, 4)
pprint(x)
pprint(x.reshape(4, 3))
rt.reshape(x, (4, 3), theme="dark")
```
