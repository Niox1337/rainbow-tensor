# Tensor shapes

`rt.shape` draws the structure of a tensor. Every axis gets its own colour from the rainbow ramp, and the leaf axis stays neutral.

```python
import numpy as np
import rainbow_tensor as rt
```

## From a vector to a high rank tensor

The same call scales from one axis up to many.

```python
rt.shape(np.arange(3))
```

```python
rt.shape(np.arange(6).reshape(2, 3))
```

```python
rt.shape(np.arange(8).reshape(2, 2, 2))
```

```python
rt.shape(np.arange(16).reshape(2, 2, 2, 2))
```

## Float precision

Floats format to a chosen number of decimals.

```python
rt.shape(np.linspace(0, 1, 6).reshape(2, 3), precision=3)
```

## Wide axes truncate

A long axis keeps a head and a tail with an ellipsis cell between them.

```python
rt.shape(np.arange(30))
```

## Save to a file

Every result carries its SVG and can write it straight to disk.

```python
visual = rt.shape(np.arange(8).reshape(2, 2, 2))
visual.save("shape_2x2x2.svg")
visual
```
