# Einsum visual

The repeated `j` label is the part to watch here. It appears in both operands and disappears from the output, so the view marks the row and column that feed the first output cell.

```python
import numpy as np

import rainbow_tensor as rt

a = np.arange(6).reshape(2, 3)
b = np.arange(12).reshape(3, 4)
```

```python
rt.einsum("ij,jk->ik", a, b)
```

The same renderer handles more than two operands. Shape tuples are useful when you want the diagram without caring about real values.

```python
rt.einsum("abc,cde,ef->abdf", (2, 2, 2), (2, 2, 2), (2, 2))
```
