# Backend arrays

rainbow-tensor reads array values through a tiny duck typed adapter. NumPy works as before, and Torch or JAX arrays render the same way when those packages are installed.

```python
import importlib.util

import numpy as np

import rainbow_tensor as rt
```

```python
x = np.arange(6).reshape(2, 3)
rt.shape(x)
```

```python
if importlib.util.find_spec("torch") is not None:
    import torch

    torch_visual = rt.shape(torch.arange(6).reshape(2, 3))
else:
    torch_visual = rt.shape(x)
torch_visual
```

```python
if importlib.util.find_spec("jax") is not None:
    import jax.numpy as jnp

    jax_visual = rt.shape(jnp.arange(6).reshape(2, 3))
else:
    jax_visual = rt.shape(x)
jax_visual
```
