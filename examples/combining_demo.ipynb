# Combining tensors

Visualise the operations that join several tensors into one. Each operand is drawn in its own tint, and the result colours every cell by the operand it came from, so the seam between operands stays clear.

```python
from pprint import pprint

import numpy as np

import rainbow_tensor as rt
```

## Concatenate

`concatenate` joins operands along an existing axis. The operands must match on every axis except the joined one, which grows by adding their sizes. Here two `(2, 3)` tensors join along axis 0 into a `(4, 3)` result.

```python
a = np.arange(6).reshape(2, 3)
b = np.arange(100, 106).reshape(2, 3)
pprint(a)
pprint(b)
pprint(np.concatenate([a, b], axis=0))
rt.concatenate([a, b], 0)
```

Joining along axis 1 grows the columns instead, and the operands may have different sizes along that axis.

```python
a = np.arange(4).reshape(2, 2)
b = np.arange(100, 106).reshape(2, 3)
pprint(a)
pprint(b)
pprint(np.concatenate([a, b], axis=1))
rt.concatenate([a, b], 1)
```

## Stack

`stack` places the operands onto a brand new axis, so every operand keeps its full shape and the result gains one rank. Stacking two `(2, 3)` tensors on axis 0 makes a `(2, 2, 3)` result.

```python
a = np.arange(6).reshape(2, 3)
b = np.arange(100, 106).reshape(2, 3)
pprint(a)
pprint(b)
pprint(np.stack([a, b], axis=0))
rt.stack([a, b], 0)
```

The new axis can sit anywhere. On axis 2 it becomes the innermost axis.

```python
a = np.arange(6).reshape(2, 3)
b = np.arange(100, 106).reshape(2, 3)
pprint(a)
pprint(b)
pprint(np.stack([a, b], axis=2))
rt.stack([a, b], 2)
```

## Mismatched shapes

When the operands do not line up for the chosen axis the call raises a clear error instead of drawing a wrong figure.

```python
a = np.zeros((2, 3))
b = np.zeros((2, 4))
try:
    rt.concatenate([a, b], 0)
except ValueError as error:
    print(error)
```

## Dark theme

The tints adapt to the dark theme through `theme="dark"`.

```python
a = np.arange(6).reshape(2, 3)
b = np.arange(100, 106).reshape(2, 3)
pprint(np.concatenate([a, b], axis=0))
rt.concatenate([a, b], 0, theme="dark")
```
