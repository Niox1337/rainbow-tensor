# Reshaping and moving axes

These views rearrange a tensor without changing its values. Each axis keeps its colour so you can follow where it lands.

```python
import numpy as np

import rainbow_tensor as rt
```

## Reshape

Reshape keeps the row major order, so element k stays element k. A single `-1` infers one axis from the rest.

```python
rt.reshape(np.arange(6).reshape(2, 3), (3, 2))
```

```python
rt.reshape(np.arange(24).reshape(2, 3, 4), (-1, 4))
```

## Transpose and swapaxes

`transpose` reverses or permutes the axes. `swapaxes` exchanges just two of them.

```python
rt.transpose(np.arange(24).reshape(2, 3, 4))
```

```python
rt.swapaxes(np.arange(24).reshape(2, 3, 4), 0, 2)
```

## Moveaxis

Move one axis to a new position and slide the others along to fill the gap.

```python
rt.moveaxis(np.arange(24).reshape(2, 3, 4), 0, 2)
```

## Squeeze and expand_dims

Drop size one axes or insert new ones. The surviving axes keep their colour and the changed axes are marked.

```python
rt.squeeze(np.ones((1, 3, 1)))
```

```python
rt.expand_dims(np.arange(6).reshape(2, 3), 1)
```
