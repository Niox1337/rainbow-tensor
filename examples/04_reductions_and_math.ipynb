# Reductions and math

These views collapse or contract axes. The source and the result sit side by side so the reduced axis is easy to see.

```python
import numpy as np

import rainbow_tensor as rt
```

## Sum and mean

A reduction collapses one axis. Values that fold into the same result share a background colour, and every surviving axis keeps its original colour.

```python
rt.sum(np.arange(24).reshape(2, 3, 4), 0)
```

```python
rt.mean(np.arange(24).reshape(2, 3, 4), 1)
```

## Matmul

Matrix multiply contracts the shared inner axis and keeps the outer axes.

```python
rt.matmul(np.arange(6).reshape(2, 3), np.ones((3, 2)))
```

## Einsum

A repeated label is the contracted axis. The same renderer handles many operands, and shape tuples work when the values do not matter.

```python
rt.einsum('ij,jk->ik', np.arange(6).reshape(2, 3), np.arange(6).reshape(3, 2))
```

```python
rt.einsum('abc,cde,ef->abdf', (2, 2, 2), (2, 2, 2), (2, 2))
```
