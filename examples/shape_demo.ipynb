# Shape visualisation

Visualise the structure of a tensor shape with `show_shape`.

```python
import rainbow_tensor as rt
```

```python
rt.show_shape((3,))
```

```python
rt.show_shape((2, 3))
```

```python
rt.show_shape((2, 2, 2))
```

A NumPy array works too, and its real values are shown.

```python
import numpy as np

x = np.arange(8).reshape(2, 2, 2)
rt.show_shape(x)
```
