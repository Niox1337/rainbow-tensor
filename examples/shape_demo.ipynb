# Shape visualisation

Visualise the structure of a tensor shape with `shape`.

```python
import numpy as np
import rainbow_tensor as rt
```

```python
rt.shape(np.arange(3))
```

```python
rt.shape(np.arange(6).reshape(2, 3))
```

```python
rt.shape(np.arange(8).reshape(2, 2, 2))
```

## Themes

Every view also renders in a dark theme. Pass `theme="dark"` to a single call, or set a module default with `rt.set_default_theme`.

```python
rt.shape(np.arange(8).reshape(2, 2, 2), theme="dark")
```

## Float precision

Floats are formatted to a configurable number of decimals and the numbers are right aligned so a column lines up.

```python
x = np.linspace(0, 1, 6).reshape(2, 3)
rt.shape(x, precision=3)
```

## A wide tensor truncates

A long axis keeps a head and a tail with an ellipsis cell in between, so a wide tensor stays readable instead of overflowing.

```python
rt.shape(np.arange(30))
```

## A custom theme

A theme is a plain object. Derive a tweaked copy with `variant`, for example roomier cells, and pass it to any call.

```python
roomy = rt.LIGHT.variant(cell_w=64, cell_h=46)
rt.shape(np.arange(8).reshape(2, 4), theme=roomy)
```

## Saving to a file

Every result carries its SVG and can write it straight to disk.

```python
visual = rt.shape(np.arange(8).reshape(2, 2, 2))
visual.save("shape_2x2x2.svg")
```
