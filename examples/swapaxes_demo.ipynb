# Swapaxes visual

`swapaxes` is the two axis version of transpose. The result keeps the moved axis colours, which makes the swap easy to trace in a 3D shape.

```python
import numpy as np
import rainbow_tensor as rt

x = np.arange(24).reshape(2, 3, 4)
```

```python
rt.swapaxes(x, 0, 2)
```

Negative axes resolve the same way as NumPy.

```python
rt.swapaxes((2, 3, 4), -1, 0)
```
