# Combining tensors

These views join several tensors into one. Each operand keeps its own tint so you can see where its values land.

```python
import numpy as np
import rainbow_tensor as rt
```

## Concatenate

Concatenate joins operands along an existing axis. The sizes may differ on that axis.

```python
a = np.arange(6).reshape(2, 3)
b = np.arange(100, 106).reshape(2, 3)
rt.concatenate([a, b], 0)
```

## Stack

Stack places the operands onto a brand new axis, so every operand keeps its full shape.

```python
a = np.arange(6).reshape(2, 3)
b = np.arange(100, 106).reshape(2, 3)
rt.stack([a, b], 0)
```

## Broadcast

Broadcasting stretches a smaller operand to match a larger one without copying data.

```python
rt.broadcast((3, 1), (1, 4))
```

## Repeat and take

Repeat duplicates elements along an axis. Take gathers chosen indices along an axis.

```python
rt.repeat(np.arange(6).reshape(2, 3), 2, axis=0)
```

```python
rt.take(np.arange(12).reshape(3, 4), [0, 2, 2], axis=1)
```
